```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## Reading images

We can show images here by two ways - 
* Using opencv command - **cv.imshow** - This is not working on this environment. Therefore the next method comes into play.
* Using matplotlib command - **plt.imshow** - This seems to be working just fine. Therefore I will use this method for the rest of the tutorial examples

```python
img = cv.imread("Photos/cat.jpg")

# fig,ax = plt.subplots(1,1,figsize=(4,3))
# plt.subplots_adjust()

# ax.imshow(img)
plt.imshow(img)
plt.show()
```

### Converting the color channels

```python
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

### Finding out the properties of the read image

```python
print('Shape of the image = ',img.shape)
print('Type of the image variable = ',type(img))
```

```python
img_large = cv.imread('Photos/cat_large.jpg')

plt.imshow(img_large)
plt.show()
```

*Comment* - It seems like matplotlib scales the images so that they can be fitted to the screen the user is working on right now. This does not happen with **cv.imshow()** command in opencv. For using cv.imshow() command, you will have to rescale the image properly.
