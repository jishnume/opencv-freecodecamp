```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Creating a blank image

```python
# Actual image
img = cv.cvtColor(cv.imread("Photos/cat.jpg"),cv.COLOR_BGR2RGB)

# A blank imaege
img_blank = np.zeros((img.shape[0],img.shape[1],3),dtype='uint8')

fig,ax = plt.subplots(1,2,figsize=(15,6))
plt.subplots_adjust()

ax[0].imshow(img)
ax[0].set_title('Original Image',fontsize=18,color='red')
ax[0].axis('off')

ax[1].imshow(img_blank)
ax[1].set_title('Blank image',fontsize=18,color='red')
ax[1].axis('off')

plt.show()
```

Now, we will use the blank image to draw stuffs on it using various commands

### Paint the image with a color

```python
# Creating a red colored image
img_red = np.zeros((img.shape[1],img.shape[0],3),dtype='uint8')
img_red[:,:,0] = 255

# Creating a green colored image
img_green = np.zeros((img.shape[1],img.shape[0],3),dtype='uint8')
img_green[:,:,1] = 255

# Creating a blue colored image
img_blue = np.zeros((img.shape[1],img.shape[0],3),dtype='uint8')
img_blue[:,:,2] = 255

fig,ax = plt.subplots(1,3,figsize=(20,5))
ax[0].imshow(img_red)
ax[0].axis('off')
ax[0].set_title('Red colored Object',fontsize=18,color='black')

ax[1].imshow(img_green)
ax[1].axis('off')
ax[1].set_title('Green colored Object',fontsize=18,color='black')

ax[2].imshow(img_blue)
ax[2].axis('off')
ax[2].set_title('Blue colored Object',fontsize=18,color='black')

plt.show()
```

### Creating a muliply-colored image

```python
n_row=n_col=200
img_mult = np.zeros((100,100,3),dtype='uint8')

img_mult[:,:,0] = 100
img_mult[:,:,1] = 255
img_mult[:,:,2] = 100
img_mult[20:60,20:60,2] = 255

fig,ax = plt.subplots(1,1,figsize=(4,4))
ax.imshow(img_mult)
ax.set_title('A multi-colored image',fontsize=14,color='black')
ax.axis('off')
plt.show()
```

### Drawing a rectangle

```python
img_rect = cv.rectangle(img_blank,(10,10),(300,200),color=(0,0,255),thickness=2)

fig,ax = plt.subplots(1,1,figsize=(4,3))
ax.imshow(img_rect)
ax.axis('off')
ax.set_title('A plain rectangle',fontsize=14,color='red')
plt.show()
```

```python
img_blank1 = np.zeros((100,100,3),dtype='uint8')
img_blank1[:] = (255,255,25)
img_rect1 = cv.rectangle(img_blank1,(10,10),(50,50),color=(0,100,100),thickness=cv.FILLED)

fig,ax = plt.subplots(1,1,figsize=(4,3))
ax.imshow(img_rect1)
ax.axis('off')
ax.set_title('A filled rectangle',fontsize=14,color='red')
plt.show()
```

### Drawing a circle

```python
img_blank_not_filled = np.zeros((500,500,3),dtype='uint8')
img_blank_not_filled[:] = (0,0,0)

img_blank_filled = np.zeros((500,500,3),dtype='uint8')
img_blank_filled[:] = (0,0,0)

img_circle_not_filled = cv.circle(img_blank_not_filled,(250,250),100,color=(0,255,0),thickness=2)
img_circle_filled = cv.circle(img_blank_filled,(250,250),100,color=(0,255,0),thickness=-1)

fig,ax = plt.subplots(1,2,figsize=(8,6))

ax[0].imshow(img_circle_not_filled)
ax[0].set_title('A plain circle',fontsize=14,color='red')
ax[0].axis('off')

ax[1].imshow(img_circle_filled)
ax[1].set_title('A filled circle',fontsize=14,color='red')
ax[1].axis('off')

plt.show()
```

### Drawing a line

```python
img_blank_line = np.zeros((500,500,3),dtype='uint8')

img_line = cv.line(img_blank_line,(50,200),(400,350),color=(0,255,0),thickness=2)
fig,ax = plt.subplots(1,1,figsize=(8,6))
ax.imshow(img_line)
ax.set_title('A straight line',fontsize=14,color='red')
ax.axis('off')
plt.show()
```
