```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Reading video files

```python
# Creating a capture variable to capture the video frames
capture = cv.VideoCapture('Videos/dog.mp4')

is_true, frame = capture.read(20)
print('type of frame variable = ',type(frame))
print('shape of the frame variable = ',frame.shape)
print('is_true value = ',is_true)

# Converting color channel
frame_rgb = cv.cvtColor(frame,cv.COLOR_BGR2RGB)

# Displaying a single frame
fig,ax = plt.subplots(1,2,figsize=(15,6))
plt.subplots_adjust()

ax[0].imshow(frame)
ax[0].set_title("Dog (original colorspace)",fontsize=18,color='red')
ax[0].axis('off')

ax[1].imshow(frame_rgb)
ax[1].set_title("Dog (RGB colorspace)",fontsize=18,color='red')
ax[1].axis('off')

plt.show()
```

## Inspecting properties of the frames

```python
# No of frames
n_frame = int(capture.get(cv.CAP_PROP_FRAME_COUNT))
print('No of frames = ',n_frame)

# Frame height
frame_height = int(capture.get(cv.CAP_PROP_FRAME_HEIGHT))
print('Frame height = '+str(frame_height)+' pixels')

# Frame width
frame_width = int(capture.get(cv.CAP_PROP_FRAME_WIDTH))
print('Frame width = '+str(frame_width)+' pixels')

# fps of the video
fps = round(capture.get(cv.CAP_PROP_FPS),4)
print('fps of the video = '+str(fps)+' frames/sec')
```
