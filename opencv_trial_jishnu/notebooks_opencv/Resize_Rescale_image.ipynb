```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Rescale function

```python
# frame : this is the individual frame from a video file
# scale : this is the amount by which we would like to rescale our images

def rescale_image(frame,scale):
    height_original = int(frame.shape[0])
    width_original = int(frame.shape[1])

    height_scaled = int(scale*height_original)
    width_scaled = int(scale*width_original)

    dim_original = (width_original,height_original)
    dim_scaled = (width_scaled,height_scaled)

    return cv.resize(frame,dim_scaled,interpolation=cv.INTER_AREA)
```

### Rescaling an image

```python
# Reading an image
img = cv.imread('Photos/cat.jpg')
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)

print('Original (width,height) : ('+str(img_rgb.shape[1])+','+str(img_rgb.shape[0])+') (pixels)')
scale_value = 0.75

# rescale image
img_rescale = rescale_image(img_rgb,scale_value)
print('Scaled (width,height) : ('+str(img_rescale.shape[1])+','+str(img_rescale.shape[0])+') (pixels)')

fig,ax = plt.subplots(1,2,figsize=(15,6))
plt.subplots_adjust()

ax[0].imshow(img_rgb)
ax[0].axis('off')
ax[0].set_title('Original Image',fontsize=18,color='red')

ax[1].imshow(img_rescale)
ax[1].axis('off')
ax[1].set_title('Rescaled image',fontsize=18,color='red')

plt.show()
```
